Pandas Data Manipulation using agg()

```python
import pandas as pd
import numpy as np
```

```python
# Loading data
performance = pd.read_csv('C:\\Users\\user\\datasets\\StudentsPerformance.csv')
performance.head()
```

```python
race_gender_math_writing_reading = performance[["gender", "race/ethnicity", "math score", "reading score", "writing score"]]
race_gender_math_writing_reading
```

```python
race_gender_math_writing_reading.groupby(["race/ethnicity","gender"]).agg([min, max, sum])
```
